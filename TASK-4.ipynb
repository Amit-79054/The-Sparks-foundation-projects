## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Importing the dataset and creating the Dataframe

```python
import sklearn.datasets as datasets
iris=datasets.load_iris()

df=pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species']=iris.target
df.head(5)
```

# Exploratory analysis

```python
sns.set_style('darkgrid')
sns.pairplot(df,hue='Species',palette='rocket')
```

```python
sns.countplot(x='Species',data=df)
```

# Spliting the data into training and test datasets

```python
from sklearn.model_selection import train_test_split
X=df.drop('Species',axis=1)
y=df['Species']
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3)
```

# Training the model with the training datasets and predicting the test dataset

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
prediction=dtc.predict(X_test)
```

# Checking the models performance using Confusion matrix and Classification report

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,prediction))
print('\n')
print(classification_report(y_test,prediction))
```

### The model shows  98% accuracy on the test dataset.
